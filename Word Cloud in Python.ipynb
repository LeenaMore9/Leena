# Word Cloud in Python

### Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. Significant textual data points can be highlighted using a word cloud. Word clouds are widely used for analyzing data from social network websites.

## Example 1

Import Some libraries as follows,

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from wordcloud import WordCloud,STOPWORDS
from PIL import Image
```

Open blank image on which you want to perform wordcloud

```python
image=Image.open("alice_mask.png")
image
```

```python
alice_mask=np.array(Image.open("alice_mask.png"))
alice_mask
```

Create function to convert blank(0) image to color(255)image array

```python
def transform_format(val):
    if val==0:
        return 255
    else:
        return val
```

Data you want to insert in Image, that data you have to take in one variable.

```python
new= open ("alice_novel.txt","r") 
text=new.read()
```

```python
transformed_alice_mask=np.ndarray((alice_mask.shape[0],alice_mask.shape[1]),np.int32)
for i in range(len(alice_mask)):
    transformed_alice_mask[i]=list(map(transform_format,alice_mask[i]))
transformed_alice_mask
```

```python
wc=WordCloud(background_color="white",max_words=1000,mask=transformed_alice_mask,
    contour_width=3,contour_color='firebrick').generate(text)
```

```python
plt.figure(figsize=[20,10])
plt.imshow(wc,interpolation='bilinear')
plt.axis("off")
plt.show()
```

## Example 2

```python
df=pd.read_csv('winemag.csv')
df.head()
```

```python
text=df.description[0]
stopwords=set(STOPWORDS)
stopwords.update(["drink","now","wine","flavor","flavors"])
```

```python
image=Image.open("wine_mask.png")
image
```

```python
wine_mask=np.array(Image.open("wine_mask.png"))
wine_mask
```

```python
transformed_wine_mask=np.ndarray((wine_mask.shape[0],wine_mask.shape[1]),np.int32)
for i in range(len(wine_mask)):
    transformed_wine_mask[i]=list(map(transform_format,wine_mask[i]))
transformed_wine_mask
```

```python
wc=WordCloud(background_color="white",max_words=1000,mask=transformed_wine_mask,
    stopwords=stopwords,contour_width=3,contour_color='firebrick').generate(text)
```

```python
plt.figure(figsize=[20,10])
plt.imshow(wc,interpolation='bilinear')
plt.axis("off")
plt.show()                                                  
```
